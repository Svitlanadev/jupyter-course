```python
#import the packages "Pandas", "Numpy" and "MatPlotLib" into Jupyter Notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import apple's stock data
apple = pd.read_csv('apple.csv')
```

```python
apple.index
```

```python
# set column "Date" as index
apple.set_index('Date', inplace=True)
```

```python
apple.index
```

```python
#visualisation of Close price with method loc[] in 2016,2015 years
apple.loc['2016-12-31':'2016-01-01', 'Close'].plot(color="brown", label="2016")
apple.loc['2015-12-31':'2015-01-01', 'Close'].plot(color="blue", label="2015")
plt.legend()
```

```python
#Create the new column in the DataFrame using .rolling() - Moving Average
#40days
apple['MA40']=apple['Close'].rolling(40).mean()
#200 days
apple['MA200']=apple['Close'].rolling(200).mean()
```

```python
#visualisation of Close price for 2015 year and Moving Average over 40 days and 200 days
apple['Close'].loc['2015-12-31':'2015-01-01'].plot(label="Close")
apple['MA40'].loc['2015-12-31':'2015-01-01'].plot()
apple['MA200'].loc['2015-12-31':'2015-01-01'].plot()
plt.legend()
```
