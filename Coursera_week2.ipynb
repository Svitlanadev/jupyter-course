```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
%matplotlib inline
```

```python
die = pd.DataFrame([1,2,3,4,5,6])
trail = 50
results = [die.sample(2, replace=True).sum().loc[0] for i in range(trail)]
```

```python
results[:10]
```

```python
freq = pd.DataFrame(results)[0].value_counts()
freq_sort = freq.sort_index()
print(freq_sort)
```

```python
freq_sort.plot(kind="bar", color="blue", figsize=(10, 4))
```

```python
relative_freq = freq_sort/trail
```

```python
relative_freq
```
