```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame()
data['Population'] = [47, 48, 85, 20, 19, 13, 72, 16, 50, 60]
```

```python
data
```

```python
data_sample_with_replacement = data.sample(5, replace=True)
data_sample_with_replacement
```

```python
data_sample_without_replacement = data.sample(5, replace=False)
data_sample_without_replacement
```

# Parameters and Statistics

```python
# Calculate mean and variance

population_mean = data.mean()
population_var = data.var(ddof=0)
print('Population mean is ', population_mean)
print('Population variance is', population_var)
```

```python

population_mean = data.mean()
population_var = data.var()
print('Population mean is ', population_mean)
print('Population variance is', population_var)
```
